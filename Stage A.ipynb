#### Merging in Pandas with the use of merge() function
#### - inner, right, outer and left

```python
fuel_df1 = fuel_data.iloc[0:19000].reset_index(drop=True)
fuel_df2 = fuel_data.iloc[19000: ].reset_index(drop=True)
```

```python
assert len(fuel_data) == (len(fuel_df1) + len(fuel_df2))

#code above checks whether the length are the same and returns an error if not
```

```python
#mofidied code above to throw a message rather than an error

try:
    assert len(fuel_data) == (len(fuel_df1) + len(fuel_df2))
except:
    print("Mismatch with length of dataframes")
```

```python
#inner merge: rows that do not match in both dataframes will be lost
pd.merge(fuel_df1, fuel_df2, how='inner')
```

```python
#outer merge: all rows in both dataframes are returned
pd.merge(fuel_df1, fuel_df2, how='outer')
```

```python
'''
trying the left which removes the right dataframe that does not match 
with the left dataframe and in this case will be none
'''
pd.merge(fuel_df1, fuel_df2, how='left')
```

```python
#trying the right which does the same as left but the opposite
pd.merge(fuel_df1, fuel_df2, how='right')
```

## Data Visualisation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(7,4))
plt.xticks(rotation=90)
fuel_unit = pd.DataFrame({'unit': ['BBL', 'GAL', 'GRAMSU', 'KGU', 'MCF', 
                                   'MMBTU', 'MWDTH', 'MWHTH', 'TON'
                                  ],
                          'count': [7998, 84, 464, 110, 11354, 180, 95, 100, 
                                   8954
                                   ]
                         })
sns.barplot(data=fuel_unit, x='unit', y='count')
plt.xlabel('Fuel Unit')
```

#### the y-axis will now be converted in the logarithmic scale cause of the large range of values

```python
g = sns.barplot(data=fuel_unit, x='unit', y='count')
g.set_yscale("log")
g.set_ylim(1,12000)
plt.xlabel('Fuel Unit')
```

```python
#pick a sample from the dataset given
sample_df = fuel_data.sample(n=50, random_state=4)
```

```python
sns.regplot(x=sample_df['utility_id_ferc1'], y=sample_df["fuel_cost_per_mmbtu"], fit_reg=False)
```

```python
#plot a boxplot
sns.boxplot(x='fuel_type_code_pudl', y="utility_id_ferc1",
            palette=["m", "g"], data=fuel_data)
```
